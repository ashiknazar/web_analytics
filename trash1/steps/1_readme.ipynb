- inorder to get familiar with bigdata tools i am planning to remove everything and start from fresh

##### remove java

- sudo apt-get purge openjdk-* -y
- sudo apt-get purge -y openjdk-* oracle-java* default-jdk default-jre
- sudo apt-get autoremove -y
- java -version

##### remove spark,hadoop,kafka if installed manually
- sudo rm -rf /usr/local/spark
- sudo rm -rf /usr/local/hadoop
- sudo rm -rf /usr/local/kafka 
- sudo rm -rf /opt/spark
- sudo rm -rf /opt/hadoop
- sudo rm -rf /opt/kafka

##### remove installed folders
- sudo rm -rf /usr/lib/jvm/java-*
- sudo rm -rf ~/hadoop
- sudo rm -rf ~/spark
- sudo rm -rf ~/kafka

##### remove python virtual environments if you want
- rm -rf ~/venvs

##### Also remove any temporary folders if you created them

- sudo rm -rf /tmp/kafka-logs
- sudo rm -rf /tmp/hadoop-*

___

##### uninstall docker

- **stop docker servicer**
   - sudo systemctl stop docker 
   - sudo systemctl disable docker
- **uninstall docker packages**

   - sudo apt-get purge -y docker-ce docker-ce-cli containerd.io docker-compose-plugin docker.io
   - sudo apt-get autoremove -y
- **Remove leftover Docker directories**
   - sudo rm -rf /var/lib/docker
   - sudo rm -rf /var/lib/containerd
- **Optional: remove Docker CLI configs in your home directory**
   - rm -rf ~/.docker

___

##### Remove Environment Variables

- backup .bashrc before editing
- `cp ~/.bashrc ~/.bashrc.bak`

- edit

- `nano ~/.bashrc`
- remove lines like

```bash
export JAVA_HOME=/usr/lib/jvm/...
export PATH=$JAVA_HOME/bin:$PATH

export SPARK_HOME=/usr/local/spark
export PATH=$SPARK_HOME/bin:$PATH

export HADOOP_HOME=/usr/local/hadoop
export PATH=$HADOOP_HOME/bin:$PATH

export KAFKA_HOME=/usr/local/kafka
export PATH=$KAFKA_HOME/bin:$PATH
```

- save and exit and reload
- `source ~/.bashrc`

___

### Install Docker

- sudo apt update
- sudo apt install -y docker.io
- sudo apt install -y docker-compose-plugin
- sudo systemctl start docker
- sudo systemctl enable docker

##### Allow Docker without sudo

- sudo usermod -aG docker $USER
- newgrp docker
